```python
import pandas as pd
import numpy as np

nav_df = pd.read_csv("../data/raw/02_nav_history.csv")

print("Shape:", nav_df.shape)
nav_df.head()
```

```python
nav_df.isnull().sum()
```

```python
nav_df.duplicated().sum()
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])
```

```python
nav_df = nav_df.sort_values(
    ["amfi_code", "date"]
).reset_index(drop=True)

nav_df.head()
```

```python
(nav_df["nav"] <= 0).sum()
```

```python
nav_df["nav"] = nav_df.groupby("amfi_code")["nav"].ffill()
```

```python
nav_df["nav"].isnull().sum()
```

```python
nav_df.to_csv(
    "../data/processed/clean_nav.csv",
    index=False
)

print("clean_nav.csv saved successfully")
```
