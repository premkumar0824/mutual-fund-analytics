```python
import pandas as pd

fund_df = pd.read_csv("../data/raw/01_fund_master.csv")

print("Total Schemes:", fund_df["scheme_name"].nunique())

fund_df["fund_house"].value_counts()
```

```python
fund_df["category"].value_counts()
```

```python
fund_df["risk_category"].value_counts()
```

```python
import matplotlib.pyplot as plt

fund_df["category"].value_counts().plot(kind="bar")

plt.title("Fund Category Distribution")
plt.xlabel("Category")
plt.ylabel("Number of Schemes")
plt.show()
```

```python
import matplotlib.pyplot as plt

fund_df["risk_category"].value_counts().plot(kind="bar")

plt.title("Risk Category Distribution")
plt.xlabel("Risk Category")
plt.ylabel("Number of Schemes")
plt.show()
```

```python
fund_df["sub_category"].value_counts()
```

```python
import matplotlib.pyplot as plt

fund_df["sub_category"].value_counts().plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Sub Category Distribution")
plt.xlabel("Sub Category")
plt.ylabel("Number of Schemes")
plt.xticks(rotation=45)
plt.show()
```

## Sub Category Analysis

### Key Findings

- The dataset contains multiple equity and debt sub-categories.
- Large Cap, Mid Cap, Small Cap, Flexi Cap, ELSS and Value funds are represented.
- Debt-oriented categories include Liquid, Gilt and Short Duration funds.
- The dataset provides good diversification across investment styles.

### Observation

The mutual fund universe represented in this dataset covers a broad range of investment strategies and risk profiles, making it suitable for portfolio and performance analytics.

```python
fund_df["launch_date"] = pd.to_datetime(
    fund_df["launch_date"],
    errors="coerce"
)

fund_df["launch_year"] = fund_df["launch_date"].dt.year

fund_df["launch_year"].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

fund_df["launch_year"].value_counts().sort_index().plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Scheme Launch Year Distribution")
plt.xlabel("Launch Year")
plt.ylabel("Number of Schemes")
plt.show()
```

## Launch Year Analysis

### Key Findings

- Scheme launch years are distributed across multiple years.
- Both older and newer mutual fund schemes are represented.
- The dataset provides a good mix of mature and recently launched funds.

### Observation

The presence of schemes launched across different years enables future analysis of fund age, longevity, and performance consistency.

# Expense Ratio Analysis

## Key Findings

- Expense ratios vary across mutual fund schemes.
- Most schemes fall within a limited expense ratio range.
- Expense ratio is an important cost factor for investors.
- Lower expense ratios generally improve long-term returns.

## Observation

The expense ratio distribution shows the cost structure of mutual fund schemes and can be used to compare fund efficiency and investor costs.
